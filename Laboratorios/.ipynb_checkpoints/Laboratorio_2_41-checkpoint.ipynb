Esta misión consiste en proponer, implementar y evaluar el desempeño modelo(s) que busque(n) explicar las relaciones entre factores socio-demográficos y desempeño en la prueba.

El entregable de esta misión es un reporte sobre el desempeño de los modelos propuestos para abordar al menos una de las preguntas guía planteadas, acompañado de una conclusión sobre los resultados del modelo (si son válidos) en el contexto de la problemática planteada.

```python
#Definición de nuevo subconjunto de las variables para abordar misión 3 basandose en las variables categoricas que tienen mayor
#impacto sobre el puntaje global de los estudiantes en las pruebas saber 11.
#Considerando que las variables "FAMI_EDUCACIONPADRE","FAMI_EDUCACIONMADRE tienen distribución de pontaje similar por categoria
#se elimina la variable "FAMI_EDUCACIONPADRE".
#Similar situación se presenta con las variables "FAMI_TIENEINTERNET", "FAMI_TIENECOMPUTADOR", razón por la cual se elimina la 
#variable "FAMI_TIENECOMPUTADOR".

icfes2020_mod = icfes2020_new[["ESTU_GENERO", "FAMI_ESTRATOVIVIENDA", "FAMI_EDUCACIONMADRE", "FAMI_TIENEINTERNET","PUNT_GLOBAL"]]
icfes2020_mod

#Las variables a considerar en el modelo son: 
#FAMI_EDUCACIONMADRE
#FAMI_ESTRATOVIVIENDA
#FAMI_TIENEINTERNET
#ESTU_GENERO
#Se contemplan estas variables dado que brindan mayor detalle con respecto a una posible calificación de los estudiantes
#en las pruebas saber 11, de acuerdo con sus categorias.
```

```python
#Se procede a crear el dataset de variables independientes X
X=icfes2020_mod[['FAMI_EDUCACIONMADRE','FAMI_ESTRATOVIVIENDA','FAMI_TIENEINTERNET','ESTU_GENERO']]
#Reemplazar las variables categoricas por númericas para poder evaluar el modelo:
X=pd.get_dummies(data=X,drop_first=True)
X.head()
```

```python
#Creación del dataset de la variable dependiente Y
Y=icfes2020_mod["PUNT_GLOBAL"]
#División de los datos en grupo de entrenamiento y de prueba, utilizando el paquete especializado sklearn.linear_model
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.4, random_state=101)
```

```python
#Procedemos a entrenar el algoritmo usando el metodo fit() de la regresion lineal
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
```

```python
#coeficientes escogidos por el modelo de regresion
coef_df = pd.DataFrame(regressor.coef_, X.columns, columns=['Coeficiente'])
coef_df
```

```python
#Ejecución de Predicciones en los datos de prueba
Y_pred = regressor.predict(X_test)
```

```python
#Comparación de los valores de salida actuales con los valores de predicción
df_n = pd.DataFrame({'Actual': Y_test, 'Predicción': Y_pred})
df_n
```

```python
#Gráfica de Predicciones
sns.regplot(Y_test,Y_pred)
```

```python
#Reporte de medidas de desempeño relevantes para el modelo propuesto
import statsmodels.api as sm
X_train_Sm=sm.add_constant(X_train)

X_train_Sm=sm.add_constant(X_train)
ls=sm.OLS(Y_train,X_train_Sm).fit()
print(ls.summary())
```

```python
#El valor de Rsquared para el modelo aplicado, con las variables consideradas es de 20%, lo cual indica que 
#el modelo no es preciso, sin embargo, esto se puede dar porque las calificaciones de las pruebas saber 11 tienen
#inherentemente una alta cantidad de variación que no es explicable.Se podria manejar 
#el tener datos de multiples años pudo mejorar la precisión
```

```python
# #Para usar el tablero de control,se debe seleccionar en la lista desplegable la variable contra la cual desea evaluar
#la distribución del puntaje global en las pruebas saber 11
seleccion= pn.widgets.Select(name="Variable",options=icfes2020_new.columns.unique().tolist())
pn.extension('plotly')
@pn.depends(seleccion.param.value)

def interactiva(seleccion):

    figura=sns.violinplot(x='PUNT_GLOBAL',y=icfes2020_new[seleccion],data=icfes2020_new)
    return figura

tablero_control=pn.Row(pn.Column(seleccion,interactiva))
tablero_control.servable()
```
